### Changelog
- use gating to avoid overflow
- `e_layers` : 1
- `d_conv` : 전체 시퀀스의 5% (반올림)
- `d_ff` : log_2(n_class) (올림)


### Handwriting에서 유일하게 40% 성능 (57.3%)
- average 적용했을 때보다 성능이 감소하는 케이스 -> 초기 학습 시 w의 학습이 선행되었을 수 있음
- d_state를 압도적으로 축소해도 d_model만 크다면 멀쩡하게 돌아감.

- positional embedding을 빼면 미친듯이 성능이 감소함.

- timevariant_C false로 해도 성능 변화 거의 없음.

- 레이어2, dmodel 512도 54.5% 성능은 나옴

```bash
python -u ../run.py \
  --use_gpu True \
  --gpu 1 \
  --task_name classification \
  --is_training 1 \
  --data UEA \
  --root_path ../dataset/Handwriting/ \
  --model_id Handwriting \
  --model GatedMamba \
  --num_kernels 8 \
  --e_layers 1 \
  --d_model 1024 \
  --expand 1 \
  --d_ff 5 \
  --d_conv 4 \
  --batch_size 16 \
  --des 'Exp' \
  --itr 1 \
  --dropout 0.1 \
  --learning_rate 0.001 \
  --weight_decay 0.0001 \
  --train_epochs 200 \
  --patience 20
  ```
